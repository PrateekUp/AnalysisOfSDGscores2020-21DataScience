```python
#sorting the data of SDG 15 wrt SDG 13 in descending order and taking a sample of 7 data
data3 = data2.sort_values('SDG 13', ascending=False).head(7)
data3
```

```python
#applying t-test to the sample
from scipy.stats import ttest_1samp
mean01 = np.mean(data3['SDG 15'])
print(mean01)
tset, pval = ttest_1samp(data3['SDG 15'], mean1)
# print(“p-values”,pval)
if pval < 0.025 or pval>0.975:    # alpha value is 0.05 or 5%
   print(" we are rejecting null hypothesis")
else:
  print("we are accepting null hypothesis")
```
