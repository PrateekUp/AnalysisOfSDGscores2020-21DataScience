```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df=pd.read_csv('Composite Score.csv')
```

```python
#Data Cleaning
df13=pd.read_csv('SDG 13.csv')
A=df13.shape
B=df13.columns
#print(A)
#print(B)
cont = []
for i in range(A[1]-3):
    cont.append(B[i+3])
for i in cont:
    x=df13[i].mean()
    df13[i].fillna(x, inplace=True)
df13.head(10)
```

```python
#Analysis of SDG 13
#print(cont)
SDG13_min = []
for i in range(3,A[1]):
    SDG13_min.append(df13[B[i]].min(axis = 0))
#print(SDG13_min)
SDG13_max = []
for i in range(3,A[1]):
    SDG13_max.append(df13[B[i]].max(axis = 0))
#print(SDG13_max)
SDG13_min_state = []
for i in range(3,A[1]):
    for j in range(A[0]):
        if df13.iloc[j,i] == SDG13_min[i-3]:
            break;
    SDG13_min_state.append(df13.iloc[j,2])
#print(SDG13_min_state)
SDG13_max_state = []
for i in range(3,A[1]):
    for j in range(A[0]):
        if df13.iloc[j,i] == SDG13_max[i-3]:
            break;
    SDG13_max_state.append(df13.iloc[j,2])
#print(SDG13_max_state)
SDG13_avg = []
for i in range(3,A[1]):
    SDG13_avg.append(df13[B[i]].mean(axis = 0))
SDG13_value={'Topic': cont, 'State_Min': SDG13_min_state, 'Min_Value': SDG13_min, 'State_Max': SDG13_max_state,
    'Max_Value': SDG13_max, 'Average_Val': SDG13_avg}
data_13=pd.DataFrame.from_dict(SDG13_value, orient='index')
data_13=data_13.transpose()
data_13
```

```python
#Performing Shapiro Wilk Test
#The Shapiro–Wilk test tests the null hypothesis that a sample x1, ..., xn came from a normally distributed population.
from scipy.stats import shapiro
DataToTest=df["SDG 13"].tolist()
stat, p=shapiro(DataToTest)
print('stat=%.2f, p=%.30f' % (stat,p))
if p> 0.05:
    print("Yes, normal distribution")
else:
    print("Not a normal distribution")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
train=df13.drop(['SNO','Category','States/UTs'], axis=1)
test=df['SDG 13']
X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=None, train_size=None, random_state=3, shuffle=True)
regr=LinearRegression()
regr_model=regr.fit(X_train, y_train)
pred=regr.predict(X_test)
print(pred)
print(regr.score(X_test, y_test))
#print(X_test)
compare = pd.DataFrame({'Actual': y_test, 'Predicted': pred})
print(compare)

plt.scatter(y_test, pred)
plt.title("Plot of Predicted value vs Actual Value")
plt.xlabel("Actual Value")
plt.ylabel("Predicted value")
plt.show()
plt.close()
```

```python
#Support Vector Machine Model Demonstration
from sklearn.svm import SVC  
y=[]
for i in df13['Category']:
    if(i=='Union Territory'):
        y.append(1)
    else:
        y.append(0)
#print(y)
train2=train.drop([cont[0], cont[1], cont[2]], axis=1)
x = train2.to_numpy()
model = SVC(kernel='linear') 
model.fit(x, y)
#get the hyperplane
w=model.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(0,300)
yy=a*xx-model.intercept_[0]/w[1]
#Plotting the hyperplane
g=sns.lmplot(x=cont[3], y=cont[4], data=df13, hue='Category', palette='Set1', fit_reg=False, scatter_kws={'s': 70})
plt.plot(xx, yy, linewidth=2, color='black')
```
