```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.67)})
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("D:\IIT\Projects\DataScienceproject\DSpro_sustainabledev_working\SDG 12.csv")
data
```

```python
data.columns
```

```python
x=data['States/UTs']
y=data['Per capita fossil fuel consumption (in kg.)']
plt.bar(x,y,align='edge',edgecolor='black',linewidth="2")
plt.xticks(rotation=90)
plt.xlabel("States & UTs")
plt.ylabel("Per capita fossil fuel consumption (in kg.)")
```
