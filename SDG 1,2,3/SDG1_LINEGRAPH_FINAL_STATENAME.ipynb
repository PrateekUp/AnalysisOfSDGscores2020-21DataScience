```python
df2=df2.replace("Null"," ")
```

```python
df2.index[df2['Percentage of population living below the national poverty line']==" "].tolist()
```

```python
df3=df2.drop(df2.index[23])
```

```python
df3['Percentage of population living below the national poverty line'] = df3['Percentage of population living below the national poverty line'].astype(float)
```

```python
df3['SNO'] = df3['SNO'].astype(str)
```

```python

x = df3['States/UTs']
y = df3['Percentage of population living below the national poverty line']
plt.rcParams["figure.figsize"] = [14.00, 5]

plt.ylabel("Percentage of population living below the national poverty line")
plt.xlabel("State/UT Name")
plt.xticks(rotation=90)

plt.title("Percentage of population living below the national poverty line")
# Plot a simple line chart
plt.plot(x, y, 'g',)


plt.legend()
plt.show()
```
