```python
df2=df2.replace("Null"," ")
```

```python
df2['SNO'] = df2['SNO'].astype(str)
```

```python
fig=plt.figure(figsize=[5,2])
ax= fig.add_axes([0,0,3,3])
xaxis=df2['SNO']
yaxis=df2['Percentage of beneficiaries covered under National Food Security Act (NFSA),2013']
plt.xlabel("Serial Number of States/UTs",fontsize=15)
plt.ylabel("Percentage of beneficiaries covered under National Food Security Act (NFSA),2013",fontsize=15)
plt.title("Percentage of beneficiaries covered under National Food Security Act (NFSA),2013 for different States/UTs ",fontsize=20)
ax.bar(xaxis,yaxis,color="blue",width=0.6)
plt.show()
```
