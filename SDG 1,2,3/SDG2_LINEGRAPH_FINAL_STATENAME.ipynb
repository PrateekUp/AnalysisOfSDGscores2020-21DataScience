```python
df2=df2.replace("Null"," ")
```

```python
df2['SNO'] = df2['SNO'].astype(str)
```

```python
x = df2['States/UTs']
y = df2['Percentage of beneficiaries covered under National Food Security Act (NFSA),2013'] 
plt.rcParams["figure.figsize"] = [14.00, 10]

plt.ylabel("Percentage of beneficiaries covered under National Food Security Act (NFSA),2013")
plt.xlabel("State/UT Name")
plt.xticks(rotation=90)

plt.title("Percentage of beneficiaries covered under National Food Security Act (NFSA),2013")
# Plot a simple line chart
plt.plot(x, y, 'b')


plt.legend()
plt.show()
```
